```python
df = pd.read_csv("./PWB.csv")
df.head()
```

```python
df.describe()
```

```python
# Assuming data is a 2D array with your data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
scaled_data_df = pd.DataFrame(scaled_data, columns=df.columns)
scaled_data_df.head()
```
